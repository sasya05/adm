```python
import pandas as pd
import numpy as np
data =pd.read_csv("titanic_clean.csv")
summary=data.describe()
print(summary)
```

```python
print(data.describe(include=['object']))
```

```python
import matplotlib.pyplot as plt
plt.hist(data['age'].dropna(), bins=20, edgecolor='black', color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python

plt.figure(figsize=(6,4))
data['pclass'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python

plt.figure(figsize=(6, 4))
plt.boxplot(data['age'].dropna(), patch_artist=True,
boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.show()
```

```python
import seaborn as sns
sns.pairplot(data[['age', 'fare', 'sibsp', 'parch']].dropna())
plt.show()
```

```python
correlation_matrix = data[['age', 'fare', 'sibsp', 'parch']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='survived', data=data,hue='survived')
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
sns.violinplot(x='pclass', y='age', data=data,hue='pclass',palette='muted')
plt.title('Age Distribution by Passenger Class')
plt.show()
```
